```python
import pickle
import pandas as pd

df_raw2 = pd.read_pickle('sah_raw_data_batch2.pkl')
df_raw1 = pd.read_pickle('sah_raw_data.pkl')
df_raw = pd.concat([df_raw1,df_raw2])
df_raw
```

```python
df1 = pd.read_pickle('ethnic_batch_1.pkl')
df2 = pd.read_pickle('ethnic_batch_2.pkl')
df_ethnic = pd.concat([df1,df2])
df_ethnic 
```

```python
df=df_raw[['tweet_id','text']]
df
```

```python
new_df = pd.merge(df, df_ethnic ,  how='left', on=['tweet_id'])
new_df
```

```python
new_df['ethnic_inferred'].value_counts()
```

```python
df_freq = pd.read_csv('retweeted_tweet_at_least_20.csv')
df_freq =df_freq [['text']]
df_freq 
```

```python
merge_df  = pd.merge(df_freq, new_df,  how='left', on=['text'], sort=False)
merge_df
```

```python
gb = merge_df.groupby(['text','ethnic_inferred'],sort=False).count()
gb
```

```python
gb.to_csv('retweeted_tweet_at_least_20_race.csv')
gb.to_excel('retweeted_tweet_at_least_20_race.xlsx')
```
